```python
import pandas as pd 
```

```python
df=pd.read_csv('data_science_job.csv')
```

```python
df.info()
```

```python
#Handle Missing Values 
df.isnull().sum()
```

```python
df.drop(columns=['enrollee_id'],inplace=True)
```

```python
df["city"]=df["city"].apply(lambda x:x.lstrip('city_')) 
```

```python
df['city']=df['city'].astype(int)
```

```python
df['city'].dtype
```

```python
df=df.sort_values(by='city')
```

```python
df
```

```python
df['city_development_index'].fillna(method='ffill',inplace=True)
```

```python
df['gender'].fillna('Other',inplace=True)
```

```python
df['enrolled_university'].fillna('no_enrollment',inplace=True)
```

```python
df['education_level'].fillna(df['education_level'].mode()[0],inplace=True)
```

```python
df['major_discipline'].fillna('Other',inplace=True)
```

```python
df['experience'].fillna(df['experience'].median(),inplace=True)
```

```python
df['company_size'].fillna(df['company_size'].mode()[0],inplace=True)
```

```python
df['company_type'].fillna(df['company_type'].mode()[0],inplace=True)
```

```python
df['training_hours'].fillna(df['training_hours'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
#Remove Duplicates 
df.duplicated().sum()
```

```python
df=df.drop_duplicates() 
```

```python
#Handle Outliers 
df.info()
```

```python
import seaborn as sns 
```

```python
sns.boxplot(df['city_development_index'])
```

```python
sns.boxplot(df['experience'])
```

```python
sns.boxplot(df['training_hours'])
```

```python
upperlimit = df['city_development_index'].mean()+3*df['city_development_index'].std()
lowerlimit = df['city_development_index'].mean()-3*df['city_development_index'].std() 

df=df.loc[(df['city_development_index']<upperlimit) & (df['city_development_index']>lowerlimit)] 
```

```python
sns.boxplot(df['city_development_index'])
```

```python
sns.boxplot(df['training_hours'])
```

```python
upperlimit = df['training_hours'].mean()+2.5*df['training_hours'].std()
lowerlimit = df['training_hours'].mean()-2.5*df['training_hours'].std() 

df=df.loc[(df['training_hours']<upperlimit) & (df['training_hours']>lowerlimit)] 
```

```python
sns.boxplot(df['training_hours'])
```

```python
#Encoding Categorical Variables 
df['gender']=df['gender'].map({'Male': 0, 'Female': 1, 'Other': 2})
```

```python
from sklearn.preprocessing import LabelEncoder 
label=LabelEncoder()
df['relevent_experience']=label.fit_transform(df['relevent_experience'])
```

```python
df['enrolled_university']=df['enrolled_university'].map({'Full time course':0, 'no_enrollment':1, 'Part time course':2})
```

```python
df['education_level']=df['education_level'].map({'Graduate':0, 'Masters':1, 'High School':2 , 'Primary School':3, 'Phd':4})
```

```python
df['major_discipline']=df['major_discipline'].map({'STEM':0, 'Other':1, 'No Major':2, 'Business Degree':3, 'Arts':4,'Humanities':5})
```

```python
df['company_type']=df['company_type'].map({'Pvt Ltd':0, 'Other':1, 'Public Sector':2, 'Funded Startup':3,
       'Early Stage Startup':4, 'NGO':5})
```

```python
df['company_size']=df['company_size'].map({'100-500':0, '50-99':1, '10000+':2, '5000-9999':3, '10/49':4, '<10':5,
       '500-999':6, '1000-4999':7})
```

```python
df
```

```python
#Feature Scaling (Standardization)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['city_development_index','experience', 'training_hours']] = scaler.fit_transform(df[['city_development_index','experience','training_hours']])
```

```python
from sklearn.model_selection import train_test_split 
x=df.iloc[:,0:-1] 
y=df.iloc[:,-1:] 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score,precision_score, recall_score, f1_score
```

```python
accuracy_score(y_test, y_pred)*100
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)


print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"ROC-AUC: {roc_auc:.2f}")
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```
